```python
# conda activate mri
import numpy as np
import pandas as pd
import pingouin as pg

# figures
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.offsetbox import AnchoredText
%matplotlib inline
```

```python
# grab data only for participants included (n=150)
banda_f = '/projects/swglab/data/BANDA_U01/analysis/proj_franci_anxiety/behave_inputs/consort_final_sample_banda_n150.csv'
banda = pd.read_csv(banda_f).set_index(['participant_id'])
banda.head(2)
```

# Correlations at T1 (baseline)

```python
x = banda['rcads01_rcads_drs_T1']
y = banda['rcads01_rcads_ALLANX_T1']
print(pg.corr(x,y, method = 'pearson', alternative = "two-sided").round(3)[['r','p-val']])
```

```python
x = banda['rcads01_rcads_drs_T3']
y = banda['rcads01_rcads_ALLANX_T3']
print(pg.corr(x,y, method = 'pearson', alternative = "two-sided").round(3)[['r','p-val']])
```

```python
t = "T1"
# anx = ['rcads_anx','rcads_anxonly','rcads_gad']
banda[['rcads01_rcads_ALLANX_T1','rcads01_rcads_ALLANX_T3']]
for a in anx:
    x = banda.loc[:,t,:]['rcads_mdd']
    y = banda.loc[:,t,:][a]
    print('\n\n' ,a)
    print(pg.corr(x,y, method = 'pearson', alternative = "two-sided").round(3)[['r','p-val']])
```

## at T3 (1-year follow-up)

```python
t = "T3"
anx = ['rcads_anx','rcads_anxonly','rcads_gad']
for a in anx:
    x = banda.loc[:,t,:]['rcads_mdd']
    y = banda.loc[:,t,:][a]
    print('\n\n' ,a)
    print(pg.corr(x,y, method = 'pearson', alternative = "two-sided").round(3)[['r','p-val']])
```

# correlations T1

```python
banda['rcads01_rcads_ALLANX_T1']=banda[['rcads01_rcads_social_phob_T1','rcads01_rcads_panic_T1', 'rcads01_rcads_gen_anx_T1','rcads01_rcads_ocd_T1','rcads01_rcads_sep_anx_T1']].sum(axis=1)
tmpbase = banda[['rcads01_rcads_tot_T1','rcads01_rcads_social_phob_T1','rcads01_rcads_panic_T1', 'rcads01_rcads_drs_T1', 'rcads01_rcads_gen_anx_T1','rcads01_rcads_ocd_T1','rcads01_rcads_sep_anx_T1','rcads01_rcads_ALLANX_T1']]

banda['rcads01_rcads_ALLANX_T3']=banda[['rcads01_rcads_social_phob_T3','rcads01_rcads_panic_T3', 'rcads01_rcads_gen_anx_T3','rcads01_rcads_ocd_T3','rcads01_rcads_sep_anx_T3']].sum(axis=1)
tmpfwu = banda[['rcads01_rcads_tot_T3','rcads01_rcads_social_phob_T3','rcads01_rcads_panic_T3', 'rcads01_rcads_drs_T3', 'rcads01_rcads_gen_anx_T3','rcads01_rcads_ocd_T3','rcads01_rcads_sep_anx_T3','rcads01_rcads_ALLANX_T3']]
```

```python
symptoms = pg.pairwise_corr(tmpbase, method = "pearson", padjust="bonf").set_index('X').round(4)#[['Y','r','p-corr']]   #columns=['rcads_tot'], 
symptoms
```

```python
symptoms.describe().round(2).loc[['mean','std','min','max']][['r','p-corr']]
```
